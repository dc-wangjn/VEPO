```python
import json
import numpy as np
def read_baseline(path,agent="IQN",iteration =200):
    f = open(path,encoding='utf-8')
    res = f.read()
    data = json.loads(res)

    returns = []
    for i in range(iteration):
        iter_data =  []
        for x in data:

            if x.get('Iteration')==i and x.get('Agent')==agent:
                iter_data.append(x.get('Value'))
        returns.append([np.mean(np.array(iter_data)),np.std(np.array(iter_data))])
    return np.array(returns)
```

```python
import matplotlib.pyplot as plt
game = 'breakout'
raw_data, _ = colab_utils.load_statistics(
  'expriment/{}/logs'.format(game), verbose=False)
summarized_data = colab_utils.summarize_data(
  raw_data, ['train_episode_returns'])

length = len(summarized_data['train_episode_returns'])
baseline = read_baseline('baselines/data/{}.json'.format(game),agent='IQN',iteration =200)

plt.plot(summarized_data['train_episode_returns'], label='TBIQN')
# plt.plot(summarized_data1['train_episode_returns'], label='TBIQNv2')

plt.plot(baseline[:length,0],label='IQN')
plt.plot()
plt.title('IQN training - {}'.format(game))
plt.xlabel('Iteration')
plt.ylabel('Return')
plt.legend()
plt.show()
```

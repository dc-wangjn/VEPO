```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
def read_data(nt,ts,gm,J,wh,seed):
    test_loss = []
    ww = []
    for k in seed:
        loss_fn = '/home/zf_project/TestMDP/experiment_script/result/testloss_nt={}ts={}gm={}J={}wh={}seed={}.pkl'.format(nt,ts,gm,J,wh,k)
        f = open(loss_fn,'rb')
        loss,w = pickle.load(f)
        f.close()
        test_loss.append(loss)
        ww.append(w)
    test_loss = np.array(test_loss)
    return test_loss,ww
```

```python
nt = 500
ts = 100
gm = 0.8
J = 4
wh =128
seed = range(5)
```

```python
import pickle
f = open('results/multi_robust_sigma2.0_delta0.1_3050.pkl','rb')
e = pickle.load(f)
f.close()
g = 0
# d = e[0][g][:,:,:,0]
d = e[0][g]
c = e[1][g]
K_iters = [2]
NN_TT = ((10,10),(20,20))
print(d.shape)
# print(c[0,:,0,1,0])

# g = 0
# d = e[0][g][:,:,:,0]
# d = e[0][g]

c = e[1][g]
cr = e[2][g]
config = e[3]

NN_TT = [[30,30],[50,50]]
pol =0
mod = 2
sam = 0
multiplier = 1
alpha = 1

import numpy as np
import matplotlib.pyplot as plt
x = range(K_iters[0]+2)
fig = plt.figure(figsize=(20, 12))
for i in range(3):
    for j in range(2):
#         for k in range(3):
#             ax = fig.add_subplot(5,5,j*5+i+1,sharey=ax)
            ax = fig.add_subplot(2,3,j*3+i+1)
#             ax = fig.add_subplot(1,3,j*3+i+1,sharey=ax)
            yy = d[:,i,0,j][range(K_iters[0]+1)]
            std = np.std(yy,axis=1)
            yy = np.mean(yy,axis=1)
            y_n1_t50 = np.hstack([yy[0,0],yy[:,1]]).reshape(-1)
            
            
            std1 = np.hstack([std[0,0],std[:,1]]).reshape(-1)
            print('row {} col {}'.format(j,i))
            print(y_n1_t50)
            print(std1)

            yy = d[:,i,1,j][range(K_iters[0]+1)]
            std = np.std(yy,axis=1)
            yy = np.mean(yy,axis=1)
            y_n2_t50 = np.hstack([yy[0,0],yy[:,1]]).reshape(-1)
            
            std2 = np.hstack([std[0,0],std[:,1]]).reshape(-1)
            print(y_n2_t50)
            print(std2)

            yy = d[:,i,2,j][range(K_iters[0]+1)]
            std = np.std(yy,axis=1)
            yy = np.mean(yy,axis=1)
            y_n3_t50 = np.hstack([yy[0,0],yy[:,1]]).reshape(-1)
            
            
            std3 = np.hstack([std[0,0],std[:,1]]).reshape(-1)
            print(y_n3_t50)
            print(std3)
            
            yy = d[:,i,3,j][range(K_iters[0]+1)]
            std = np.std(yy,axis=1)
            yy = np.mean(yy,axis=1)
            y_n4_t50 = np.hstack([yy[0,0],yy[:,1]]).reshape(-1)
    
            
            std4 = np.hstack([std[0,0],std[:,1]]).reshape(-1)
            print(y_n4_t50)
            print(std4)
            
            yy = d[:,i,4,j][range(K_iters[0]+1)]
            std = np.std(yy,axis=1)
            yy = np.mean(yy,axis=1)
            y_n5_t50 = np.hstack([yy[0,0],yy[:,1]]).reshape(-1)
            
            
            std5 = np.hstack([std[0,0],std[:,1]]).reshape(-1)
            print(y_n5_t50)
            print(std5)
            
            ax.plot(x,y_n1_t50,'-o',label='origin')
            ax.plot(x,y_n2_t50,'-o',label='mod1')
            ax.plot(x,y_n3_t50,'-o',label='mod2')
            ax.plot(x,y_n4_t50,'-o',label='mod3')
            ax.plot(x,y_n5_t50,'-o',label='mod4')
            ax.set_ylim(0,10.5)
            
            
            ax.fill_between(x,y_n1_t50- multiplier* std1/np.sqrt(100),y_n1_t50+ multiplier* std1/np.sqrt(100),color='lightcyan',alpha=alpha) 
            ax.fill_between(x,y_n2_t50- multiplier* std2/np.sqrt(100),y_n2_t50+ multiplier* std2/np.sqrt(100),color='linen',alpha=alpha)  
            ax.fill_between(x,y_n3_t50- multiplier* std3/np.sqrt(100),y_n3_t50+ multiplier* std3/np.sqrt(100),color='lightgreen',alpha=alpha)  
            ax.fill_between(x,y_n4_t50- multiplier* std4/np.sqrt(100),y_n4_t50+ multiplier* std4/np.sqrt(100),color='gainsboro',alpha=alpha)  
            ax.fill_between(x,y_n5_t50- multiplier* std5/np.sqrt(100),y_n5_t50+ multiplier* std5/np.sqrt(100),color='thistle',alpha=alpha)  
            ax.set_ylabel('Policy Value',fontsize=15)
            ax.set_xlabel('ITERATION',fontsize=15)
#             ax.fill_between(x,y2- multiplier* std2/np.sqrt(50),y2+ multiplier* std2/np.sqrt(50),color='linen',alpha=alpha) 
#             ax.fill_between(x,y3- multiplier* std3/np.sqrt(50),y3+ multiplier* std3/np.sqrt(50),color='aliceblue',alpha=1) 
#             ax.set_title('policy :{} tn :{}'.format(i,NN_TT[j]),fontsize=13)
            ax.set_xticks([0,1,2,3])
            labels = ax.get_xticklabels() + ax.get_yticklabels()
            [x.set_fontsize(15) for x in labels]
            ax.legend(loc=4)
# fig.savefig('toy_delta0.1.eps',dpi=600,format='eps')
```
